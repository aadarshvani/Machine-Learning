```python
df = pd.read_csv('https://drive.google.com/uc?id=1pPvyzMkTp_hssPaFhKszfaWvt9nWC_zk')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
lr.predict(X_test.iloc[3].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
b
```

```python
m * 10 + b
```
